```python
import pandas as pd
```

```python
file_path= "../data/Universe ESG.xlsx"
df_return = pd.read_excel(file_path, sheet_name= "Universe")
df_info = pd.read_excel(file_path, sheet_name= "Sheet2")
```

```python
# traitement du return
df_return2= df_return.iloc[1:, 3:].transpose()#.set_index(1)
df_return2.columns= df_return2.iloc[0, :]
df_return3= df_return2.iloc[2:, :]
df_return3["Date1"]= df_return3.iloc[:, 0]
df_return3["Date2"]= df_return3.iloc[:, 1]

df_return3= df_return3.iloc[:, 3:]
df_return3= df_return3.reset_index(drop= True)
df_return3.head()
```

```python
df_return4= df_return3
df_return4= df_return4.dropna(axis= 1)
# df_return4.head()
ticker_good= [i for i in df_return4.iloc[:, 0:48].columns]
ticker_good
```

```python
df_return4.to_csv("../data/return.csv", index= False, header= True)

del df_return
del df_return2
del df_return3
del df_return4
```

```python
# traitement des infos
info= ["EBITDA_TO_REVENUE", "EPS_GROWTH", "CURRENT_EV_TO_T12M_EBITDA", 
      "EBITDA_TO_INTEREST_EXPN", "TOT_DEBT_TO_EBITDA", "PE_RATIO", "DIVIDEND_INDICATED_YIELD",
      "EQY_DVD_YLD_IND"]

ticker_good= ticker_good + ["Date"]
df_info2= df_info.iloc[15:, 20:]
ticker= df_info2.iloc[0, :].values[1:]
ticker= ["Date"] + ticker.tolist()
print(df_info2.shape)

df_info2= df_info2.iloc[2:, :]
df_info2.columns= ticker
date= df_info2["Date"].iloc[1:]

for j in info:
    test= (df_info2.iloc[0,:] == j).values
    index_col= [i for i, x in enumerate(test) if x]
    df_test= df_info2
    df_test= df_test.iloc[:, test]
    df_test["Date"]= date
    df_test["Metric"]= df_test.iloc[0, 1]
    df_test= df_test.iloc[1:, :]
    ticker_good= [i for i in df_test.columns if i not in ["Date", "Metric"]]
    df_test= df_test[ticker_good]
    # on a des nan, on les remplace par la moyenne de la ligne
    df_test[ticker_good] = df_test[ticker_good].apply(lambda row: row.fillna(row.mean()), axis=1)
    df_test["Date"] = date
#     print(df_test.columns)
    df_test.to_csv("../data/{}.csv".format(j), index= False, header= True)
```
